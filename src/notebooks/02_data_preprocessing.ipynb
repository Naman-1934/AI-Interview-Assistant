```python
import pandas as pd 

# re = Regular Expression
import re
```

```python
resume_df = pd.read_csv("..//data//processed//resume_clean_base.csv")
resume_df.head()
```

```python
def clean_resume(text):

    # Remove URLs
    text = re.sub(
        r"http\s+|www\s+",
        " ",
        text
    )
    
    # Remove email addresses
    text = re.sub(
        r"\s+@\s+",
        " ",
        text
    )

    # Remove phone numbers
    text = re.sub(
        r"\+?\d\s\-]{8,}",
        " ",
        text
    )

    # Remove HTML tags
    text = re.sub(
        r"<.*?>",
        " ",
        text
    )

    # Remove special characters
    text  = re.sub(
        r"[^a-zA-Z ]",
        " ",
        text 
    )

    text = text.lower()

    # Remove extra spaces
    text = re.sub(
        r"\s+",
        " ",
        text
    )

    return text 
```

```python
resume_df["clean_resume_text"] = resume_df["Resume_str"].apply(clean_resume)
resume_df.head()
```

```python
resume_df["clean_resume_text"].str.len().describe()
```

```python
# Remove Empty Rows
resume_df = resume_df[resume_df['clean_resume_text'].str.len()>20]
resume_df.head()
```

```python

resume_df.to_csv("..//data//processed//resume_processed.csv", index=False)
```
